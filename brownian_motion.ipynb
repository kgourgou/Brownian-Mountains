# Discrete approximation to Brownian motion

```python
T = 10; # Final time
Dt = 0.5; # Time step


n = int(np.round(T/Dt))
print n

Bn = lambda x : x + np.sqrt(Dt)*randn() #Approximation to a Brownian "step"
f  = lambda x,mu : np.exp(-(x-mu)**2/np.sqrt(2*Dt))
pi = np.pi
```

```python
x = np.linspace(-3,3)
y = f(x,0);
dat = np.array([x,y])

c = np.cos(-pi/2.0)
s = np.sin(-pi/2.0)
rot = np.array([[c,-s],[s,c]]).T # rotate by 90 degrees

def rotatedNormal(mu=0):
    x = np.linspace(-5,5)
    points = np.array([x,f(x,mu)]).T
    points = np.dot(points, rot)
    return points

```

```python
previous = 0;
T = 10; # Final time
Dt = 0.01; # Time step

n = int(np.round(T/Dt))

pl.figure(figsize=(10, 10), dpi=120, facecolor='w', edgecolor='k')
figa = pl.gca()


for i in range(n/2):
    si = np.sqrt(Dt)*randn()
    Br = previous + si
    #pl.plot([(i-1)*Dt,i*Dt], [previous, Br],'ro--',linewidth=2)
    
   
    
    points = rotatedNormal(-previous)
    pl.plot(points[:,0]+i*Dt,points[:,1],'b-',alpha=0.2,linewidth=2)

    previous = Br;

    
figa.axes.get_xaxis().set_visible(False)
figa.axes.get_yaxis().set_visible(False)


    
    
#pl.savefig('brownian_style_large.png',dpi=120)
#pl.savefig('brownian_style_large_1.png',dpi=120)

    
```
